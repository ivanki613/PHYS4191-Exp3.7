```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.optimize as opt
```

```python
data=np.loadtxt('waferb_1_light.csv', delimiter=",",skiprows=2)
#data=np.loadtxt('3.7wafera2nd_test.csv', delimiter=",")
V_exp_1=data.T[0]
C_exp_1=data.T[1]

data=np.loadtxt('waferb_1_off.csv', delimiter=",",skiprows=2)
V_exp_1_off=data.T[0]
C_exp_1_off=data.T[1]

#print(V_exp)
#print("Voltage: ", V_exp_1)
#print("Capacitance: ", C_exp_1)
plt.scatter(V_exp_1,C_exp_1, s=10,label='1 on')
plt.scatter(V_exp_1_off,C_exp_1_off, s=10,label='1 off')
plt.xlabel('Voltage (V)')
plt.ylabel('Capacitance (F)')
plt.grid()
plt.legend(loc='best')
plt.show()
```
